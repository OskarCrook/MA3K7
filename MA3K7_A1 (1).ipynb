```python

def Bracelet_xy(x, y, n):
    sequence = [x, y]
    for i in range(2, n):
        sequence.append((sequence[-1] + sequence[-2]) % 10)
    return sequence

# Check if a sequence contains a loop and return the matched loop index
def find_matching_loop(sequence, loops):
    for idx, loop in enumerate(loops):
        loop = list(loop)
        for i in range(len(sequence) - len(loop) + 1):
            if sequence[i:i + len(loop)] == loop:
                return idx
    return None


def print_highlighted_sequences(n):

    COLOURS = [
        "\033[91m",  # Red for loop 1
        "\033[92m",  # Green for loop 2
        "\033[93m",  # Yellow for loop 3
        "\033[94m",  # Blue for loop 4
        "\033[95m"   # Magenta for loop 5
    ]
    RESET = "\033[0m"


    loops = [
        (1, 1),  # Loop 1 (simplified so I need less terms in the sequence)
        (5, 5, 0, 5, 5),
        (2, 2, 4, 6, 0, 6, 6, 2, 8, 0, 8, 8, 6, 4, 0, 4, 4, 8, 2, 0, 2, 2),
        (2, 6, 8, 4, 2, 6),
        (1, 3, 4, 7, 1, 8, 9, 7, 6, 3, 9, 2, 1, 3)
    ]


    for x in range(1, 10):
        for y in range(1, 10):
            sequence = Bracelet_xy(x, y, n)
            matching_loop_idx = find_matching_loop(sequence, loops)
            if matching_loop_idx is not None:
                print(f"{COLORS[matching_loop_idx]}x = {x}, y = {y}: {sequence}{RESET}")
            else:
                print(f"x = {x}, y = {y}: {sequence}")
# min length required to capture all sequences
n = 61
print_highlighted_sequences(n)
```
