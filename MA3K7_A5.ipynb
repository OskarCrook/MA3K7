```python
import numpy as np
import random
import matplotlib.pyplot as plt
from collections import Counter

def simulate_walk(n=3, start=(1, 1), target=(2, 2), trials=5000000):
    total_steps = 0
    steps_list = []
    step_counts = Counter()

    for _ in range(trials):
        steps = 0
        position = start
        while position != target:
            neighbors = get_neighbors(position[0], position[1], n)
            position = random.choice(neighbors)
            steps += 1
        total_steps += steps
        steps_list.append(steps)
        step_counts[steps] += 1



    # Estimate probability of reaching the target in n moves
    step_probabilities = {steps: count / trials for steps, count in step_counts.items()}

    return total_steps / trials, step_probabilities

# Run the simulation for 3x3 grid
expected_steps, step_probabilities = simulate_walk()
print(f"Expected steps to reach 9 from 1: {expected_steps:.2f}")

# Print estimated probabilities
print("Estimated probabilities for reaching in n moves:")
for steps, prob in sorted(step_probabilities.items()):
    print(f"P({steps}) ≈ {prob:.4f}")
```

```python
import numpy as np


def solve_expected_steps(n):

    size = n * n  # Total number of cells
    A = np.zeros((size, size))
    b = np.ones(size)  # RHS of Ax = b

    target_index = (n-1) * n + (n-1)  # Convert (n, n) to 1D index
    A[target_index, target_index] = 1  # E(n, n) = 0
    b[target_index] = 0

    # Fill the system of equations
    for i in range(n):
        for j in range(n):
            index = i * n + j  # Convert (i, j) to 1D index
            if index == target_index:
                continue  # Skip the target cell

            neighbors = get_neighbors(i, j, n)
            num_neighbors = len(neighbors)
            A[index, index] = 1  # Coefficient of E(i, j)

            for ni, nj in neighbors:
                neighbor_index = ni * n + nj
                A[index, neighbor_index] = -1 / num_neighbors  # Probability term

    # Solve Ax = b
    E_values = np.linalg.solve(A, b)

    # Extract expected steps from center
    center_index = (n//2) * n + (n//2)  # Convert center (approximately) to 1D index
    return E_values[center_index]


grid_size = 5
expected_steps = solve_expected_steps(grid_size)
print(f"Expected steps in {grid_size}x{grid_size} grid: {expected_steps:.2f}")
```

```python
import numpy as np
import matplotlib.pyplot as plt


odd_n_values = list(range(3, 70, 2))
expected_steps = [solve_expected_steps(n) for n in odd_n_values]
log_expected_steps = np.log(expected_steps)
sqrt_expected_steps = np.sqrt(expected_steps)

# Plot expected moves
plt.figure(figsize=(10, 8))

# First plot: Expected Moves
plt.subplot(2, 2, 1)
plt.plot(odd_n_values, expected_steps, marker='o', linestyle='-', color='b')
plt.xlabel("Grid Size (n x n)")
plt.ylabel("Expected Moves")
plt.title("Expected Moves to Reach Bottom Right for Odd n")
plt.grid(True)

# Second plot: Log of Expected Moves
plt.subplot(2, 2, 2)
plt.plot(odd_n_values, log_expected_steps, marker='s', linestyle='-', color='r')
plt.xlabel("Grid Size (n x n)")
plt.ylabel("Log of Expected Moves")
plt.title("Log of Expected Moves to Reach Bottom Right for Odd n")
plt.grid(True)

# Third plot: Square Root of Expected Moves
plt.subplot(2, 1, 2)
plt.plot(odd_n_values, sqrt_expected_steps, linestyle='-', color='g')
plt.xlabel("Grid Size (n x n)")
plt.ylabel("Square Root of Expected Moves")
plt.title("Square Root of Expected Moves to Reach Bottom Right for Odd n")
plt.grid(True)

plt.tight_layout()
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.optimize import curve_fit


odd_n_values = np.arange(3, 70, 2)
expected_steps = np.array([solve_expected_steps(n) for n in odd_n_values])


def quadratic(x, a, b, c):
    return a * x**2 + b * x + c

# Fit the quadratic model to the data
params, _ = curve_fit(quadratic, odd_n_values, expected_steps)
a, b, c = params

# Generate smooth curve for plotting
x_fit = np.linspace(min(odd_n_values), max(odd_n_values), 100)
y_fit = quadratic(x_fit, a, b, c)

# Plot the data and quadratic fit
plt.figure(figsize=(8, 6))
plt.scatter(odd_n_values, expected_steps, color='blue', label="Computed Expected Steps")
plt.plot(x_fit, y_fit, color='red', linestyle='--', label=f"Quadratic Fit: {a:.2f}n² + {b:.2f}n + {c:.2f}")
plt.xlabel("Grid Size (n)")
plt.ylabel("Expected Moves")
plt.title("Quadratic Relationship Between Grid Size and Expected Moves")
plt.legend()
plt.grid(True)
plt.show()


```
